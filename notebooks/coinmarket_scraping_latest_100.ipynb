```python
base_url = 'https://pro-api.coinmarketcap.com'
load_dotenv()
api_key = os.getenv('COINMARKET_API_KEY')
```

```python
listings_latest_url = url = '/v1/cryptocurrency/listings/latest?start=1&limit=100&sort=market_cap&cryptocurrency_type=all'
```

```python
# read json from coinmarketcap cryptocurrency latest listings api
headers = {'X-CMC_PRO_API_KEY': api_key}

response = requests.get(base_url + listings_latest_url, headers=headers)
response
```

```python
data = response.json()
data['data']
```

```python
df = pd.DataFrame(data['data'])
df.head()
```
